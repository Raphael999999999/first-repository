```python
# prompt: Find/create/generate data on shopper behaviour for a particular company.
# Prepare and explore your data, which may include merging datasets, dropping irrelevant data, cleaning, etc. Note that you are required to use a minimum of 3 methods of cleaning your data. I recommend that you use standardization as one.
# Create a clustering model that groups shoppers into a minimum of 2 groups, maximum of 4 groups.
# Use a scatterplot and 2 additional visualization techniques to show results.
# Add each group member as a shopper to the dataset, and take a look at their predicted groups.
# Evaluate your model, and describe what that evaluation means.
# Improve your model

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data for shopper behavior
np.random.seed(42)
n_customers = 10000
data = {
    'CustomerID': range(1, n_customers + 1),
    'PurchaseFrequency': np.random.randint(1, 20, n_customers),
    'AverageOrderValue': np.random.uniform(10, 200, n_customers),
    'WebsiteVisitFrequency': np.random.randint(5, 50, n_customers),
    'TimeSpentOnWebsite_minutes': np.random.uniform(2, 60, n_customers),
    'HasDiscountCode': np.random.choice([0, 1], n_customers, p=[0.7, 0.3]),
    'LastPurchaseDaysAgo': np.random.randint(1, 180, n_customers)
}
df = pd.DataFrame(data)

# Simulate some missing values and outliers for cleaning demonstration
df.loc[np.random.choice(df.index, 50), 'AverageOrderValue'] = np.nan
df.loc[np.random.choice(df.index, 30), 'WebsiteVisitFrequency'] = -1 # Simulate invalid data
df.loc[np.random.choice(df.index, 20), 'TimeSpentOnWebsite_minutes'] = 1000 # Simulate outlier

print("Original Data:")
print(df.head())
print("\nData Info:")
df.info()
print("\nMissing Values:")
print(df.isnull().sum())
print("\nDescriptive Statistics:")
print(df.describe())

# --- Data Preparation and Exploration ---

# 1. Handling Missing Values (Method 1: Imputation with Median)
df['AverageOrderValue'].fillna(df['AverageOrderValue'].median(), inplace=True)

# 2. Handling Invalid Values (Method 2: Replacing negative values with a reasonable value like 0 or median)
df['WebsiteVisitFrequency'] = df['WebsiteVisitFrequency'].apply(lambda x: max(x, 0))

# 3. Handling Outliers (Method 3: Capping outliers using IQR)
Q1 = df['TimeSpentOnWebsite_minutes'].quantile(0.25)
Q3 = df['TimeSpentOnWebsite_minutes'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
df['TimeSpentOnWebsite_minutes'] = df['TimeSpentOnWebsite_minutes'].apply(lambda x: min(x, upper_bound))

print("\nData after Cleaning:")
print(df.head())
print("\nMissing Values after Cleaning:")
print(df.isnull().sum())
print("\nDescriptive Statistics after Cleaning:")
print(df.describe())

# Select features for clustering and standardize
features = ['PurchaseFrequency', 'AverageOrderValue', 'WebsiteVisitFrequency', 'TimeSpentOnWebsite_minutes', 'HasDiscountCode', 'LastPurchaseDaysAgo']
X = df[features]

# Standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=features)

print("\nScaled Data:")
print(X_scaled_df.head())

# --- Clustering Model ---

# Use Elbow method to determine the optimal number of clusters (k)
inertia = []
k_range = range(2, 5) # Minimum 2, Maximum 4 clusters
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Method
plt.figure(figsize=(8, 4))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_range)
plt.grid(True)
plt.show()

# Based on the elbow plot, let's choose k=3 (you can adjust based on the plot)
optimal_k = 3 # Choose the elbow point, or pick a reasonable number within the range

# Apply KMeans with the chosen number of clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)

print("\nData with Cluster Assignments:")
print(df.head())

# --- Visualization of Results ---

# 1. Scatter Plot (using first two principle components for dimensionality reduction for visualization)
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)
principal_df = pd.DataFrame(data=principal_components, columns=['principal_component_1', 'principal_component_2'])
principal_df['Cluster'] = df['Cluster']

plt.figure(figsize=(10, 6))
sns.scatterplot(x='principal_component_1', y='principal_component_2', hue='Cluster', data=principal_df, palette='viridis', legend='full')
plt.title('Shopper Clusters (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# 2. Box Plot (comparing features across clusters)
df_melted = df.melt(id_vars='Cluster', value_vars=['PurchaseFrequency', 'AverageOrderValue', 'WebsiteVisitFrequency', 'TimeSpentOnWebsite_minutes'],
                    var_name='Feature', value_name='Value')
plt.figure(figsize=(12, 8))
sns.boxplot(x='Feature', y='Value', hue='Cluster', data=df_melted, palette='viridis')
plt.title('Feature Distribution Across Clusters')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 3. Count Plot (showing the number of customers in each cluster)
plt.figure(figsize=(6, 4))
sns.countplot(x='Cluster', data=df, palette='viridis')
plt.title('Number of Customers per Cluster')
plt.xlabel('Cluster')
plt.ylabel('Number of Customers')
plt.show()

# --- Adding New Shoppers and Predicting Groups ---

# Create data for new potential shoppers
new_shoppers_data = {
    'CustomerID': [n_customers + 1, n_customers + 2, n_customers + 3],
    'PurchaseFrequency': [15, 3, 8],
    'AverageOrderValue': [150, 30, 75],
    'WebsiteVisitFrequency': [40, 10, 25],
    'TimeSpentOnWebsite_minutes': [50, 15, 30],
    'HasDiscountCode': [1, 0, 1],
    'LastPurchaseDaysAgo': [10, 90, 45]
}
new_shoppers_df = pd.DataFrame(new_shoppers_data)

# Standardize the new shopper data using the same scaler
new_shoppers_scaled = scaler.transform(new_shoppers_df[features])

# Predict the cluster for the new shoppers
predicted_clusters = kmeans.predict(new_shoppers_scaled)
new_shoppers_df['Predicted_Cluster'] = predicted_clusters

print("\nNew Shoppers with Predicted Clusters:")
print(new_shoppers_df)

# Combine the original dataframe with the new shoppers for inspection
# Ensure column consistency before concatenation, handling 'Cluster' column in original df
new_shoppers_for_merge = new_shoppers_df.drop(columns=['Predicted_Cluster']).copy()
new_shoppers_for_merge['Cluster'] = new_shoppers_df['Predicted_Cluster']

df_with_new_shoppers = pd.concat([df, new_shoppers_for_merge], ignore_index=True)

print("\nCombined DataFrame (Original + New Shoppers):")
print(df_with_new_shoppers.tail())

# --- Model Evaluation ---

# Evaluate using Silhouette Score (measures how similar a data point is to its own cluster compared to other clusters)
# Score ranges from -1 (bad clustering) to +1 (good clustering). 0 indicates overlapping clusters.
silhouette_avg = silhouette_score(X_scaled, df['Cluster'])
print(f"\nSilhouette Score: {silhouette_avg}")

# Interpretation of Silhouette Score:
# - A score close to 1 indicates that the data points are well within their own clusters and far from other clusters.
# - A score near 0 indicates that clusters are close to each other and there might be overlapping.
# - A negative score indicates that data points have been assigned to the wrong clusters.
print("\nSilhouette Score Interpretation:")
if silhouette_avg > 0.7:
    print("The clustering is strong and well-separated.")
elif silhouette_avg > 0.5:
    print("The clustering is reasonably good.")
elif silhouette_avg > 0:
    print("The clustering is acceptable, but there might be some overlapping.")
else:
    print("The clustering is poor.")

# Evaluate using Inertia (sum of squared distances of samples to their closest cluster center)
# Lower inertia generally means better clustering, but it decreases as the number of clusters increases.
print(f"Inertia: {kmeans.inertia_}")

# Interpretation of Inertia:
# Inertia is useful for the Elbow Method. A lower value is better, but it doesn't tell you if the clusters are meaningful.
# It's highly dependent on the number of clusters.

# --- Model Improvement ---

# Possible ways to improve the model:
# 1. Feature Engineering: Create new features from existing ones (e.g., purchase frequency per month, average time between purchases).
# 2. Trying Different Clustering Algorithms: K-Means assumes spherical clusters of equal size. Other algorithms like DBSCAN, Agglomerative Clustering, or Gaussian Mixture Models might be more suitable depending on the data structure.
# 3. Optimizing the Number of Clusters (k): Re-evaluate the Elbow method, or use other metrics like the Calinski-Harabasz index or Davies-Bouldin index to find the optimal k.
# 4. Handling Outliers Differently: Instead of capping, you could remove outliers or use robust scaling methods.
# 5. Using More Data or Different Data Sources: More comprehensive data on customer interactions, demographics, etc., could lead to better segmentation.

print("\nPotential Model Improvements:")
print("- Consider trying different clustering algorithms (e.g., DBSCAN, Agglomerative Clustering).")
print("- Refine the choice of the number of clusters (k) using additional metrics.")
print("- Explore feature engineering to create more informative features.")
print("- Investigate alternative outlier handling techniques.")
print("- Obtain more diverse and comprehensive data.")

# Example of trying a different number of clusters (based on Elbow method)
# If the elbow plot suggested k=2 or k=4, you could re-run the clustering with that k:
# optimal_k_improved = 2 # Or 4, based on Elbow plot analysis
# kmeans_improved = KMeans(n_clusters=optimal_k_improved, random_state=42, n_init=10)
# df['Cluster_Improved'] = kmeans_improved.fit_predict(X_scaled)
# print(f"\nSilhouette Score with k={optimal_k_improved}: {silhouette_score(X_scaled, df['Cluster_Improved'])}")

# Example of trying Agglomerative Clustering
# from sklearn.cluster import AgglomerativeClustering
# agg_clustering = AgglomerativeClustering(n_clusters=optimal_k) # Using the same number of clusters
# df['Cluster_Agg'] = agg_clustering.fit_predict(X_scaled)
# print(f"\nSilhouette Score with Agglomerative Clustering (k={optimal_k}): {silhouette_score(X_scaled, df['Cluster_Agg'])}")
```
